```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv('/content/drive/MyDrive/Credit Card Fraud Detection/dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset['Class'].value_counts()
```

**Observations:**
The dataset is highly imbalanced.

- 0--> Normal trans.
- 1--> Fraudulent trans.

### Handling imbalance sataset.

```python
#separating the data based on class
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]
```

```python
legit.shape,fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
dataset.groupby('Class').mean()
```

#### Under Sampling

- Building a dataset which containing similar distribution of normal and fraudulant transactions.

- Num. of fraudulent transactions = 492

```python
legit_sample=legit.sample(492)
```

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X=new_df.drop('Class',axis=1)
y=new_df['Class']
```

```python
X
```

### Spliting data into test and train.

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

### Model Training

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
X_train_pred=lr.predict(X_train)
train_acc=accuracy_score(X_train_pred,y_train)
print('Accuracy on training data: ',train_acc)

X_test_pred=lr.predict(X_test)
test_acc=accuracy_score(X_test_pred,y_test)
print('Accuracy on test data: ',test_acc)
```
